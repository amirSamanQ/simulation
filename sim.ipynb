```python
def exponential(lambd):
    r = random.random()
    return -(1 / lambd) * math.log(r)


def uniform(a, b):
    r = random.random()
    return a + (b - a) * r

def Triangular(max,mean,min):
    r = random.random()
    F_c = (max - min) / (mean - min)
    if r <= F_c:
        return min + math.sqrt(r * (mean - min) * (max - min))
    else:
        return mean - math.sqrt((1 - r) * (mean - min) * (mean - max))
```
